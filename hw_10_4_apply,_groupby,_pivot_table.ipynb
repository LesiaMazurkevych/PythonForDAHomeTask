0. –Ü–º–ø–æ—Ä—Ç—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É pandas —Ç–∞ –∑—á–∏—Ç–∞–π—Ç–µ –¥–∞–Ω—ñ —É –∑–º—ñ–Ω–Ω—É `df` —Ç–∏–ø—É `pandas.DataFrame`.

```python
import pandas as pd
df = pd.read_csv(r'C:\Users\mazur\data_analysis\data\supermarket_sales.csv')
```

```python
df.head()
```

1. –î–æ—Å–ª—ñ–¥–∏–º–æ, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª —Å—É–ø–µ—Ä–º–∞—Ä–∫–µ—Ç—É ('Branch') —î –Ω–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à–∏–º. –î–ª—è —Ü—å–æ–≥–æ –∑–Ω–∞–π–¥—ñ—Ç—å —Å—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –∑–∞ –∫–æ–∂–Ω–∏–º —Ñ—ñ–ª—ñ–∞–ª–æ–º —ñ –≤–∏—è–≤—ñ—Ç—å, —è–∫–∏–π —Ñ—ñ–ª—ñ–∞–ª –º–∞—î –Ω–∞–π–≤–∏—â–∏–π.

```python
grouped = df.groupby('Branch')['gross income'].sum()
print(grouped)
max_branch = grouped.idxmax()
max = grouped.max()
print(f"Max income in in branch {max_branch} : {max}")
```

2. –í —è–∫–æ–º—É –º—ñ—Å—Ç—ñ –∑–Ω–∞–π—Ö–æ–¥–∏—Ç—å—Å—è —Ñ—ñ–ª—ñ–∞–ª –∑ –Ω–∞–π–≤–∏—â–∏–º –ø—Ä–∏–±—É—Ç–∫–æ–º? –ú–æ–∂–µ –≤ —Ç–æ–º—É –º—ñ—Å—Ü—ñ –Ω–∞–º —Ä–æ–∑–º—ñ—Ç–∏—Å—Ç–∏ —â–µ –æ–¥–∏–Ω –º–∞–≥–∞–∑–∏–Ω.  
–ó–Ω–∞–π–¥—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥—å –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É Pandas.

```python
df[df['Branch'] == max_branch]['City']
```

3.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –∑–≤–æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—é, —è–∫–∞ –ø–æ–∫–∞–∂–µ, —Å–∫—ñ–ª—å–∫–∏ –ø–æ–∫—É–ø–æ–∫ (—ñ–Ω–≤–æ–π—Å—ñ–≤) –±—É–ª–æ –∑—Ä–æ–±–ª–µ–Ω–æ –≤ –∫–æ–∂–Ω—ñ–π –∑ —Ñ—ñ–ª—ñ–π (`Branch`) –∑–∞ —Ä—ñ–∑–Ω–∏–º–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è–º–∏ —Ç–æ–≤–∞—Ä—ñ–≤. –ó–∞–ø–∏—à—ñ—Ç—å —Ç–∞–±–ª–∏—Ü—é –≤ –∑–º—ñ–Ω–Ω—É `invoices_by_category` —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –∑–º—ñ–Ω–Ω—É –Ω–∞ –µ–∫—Ä–∞–Ω.
–¶—è —Ç–∞–±–ª–∏—Ü—è –¥–æ–ø–æ–º–æ–∂–µ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, –≤ —è–∫—ñ–π —Ñ—ñ–ª—ñ—ó –∫—É–ø—É—é—Ç—å –Ω–∞–π–±—ñ–ª—å—à–µ —Ç–æ–≤–∞—Ä—ñ–≤ –∫–æ–∂–Ω–æ—ó –∑ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π.

```python
invoices_by_category = df.pivot_table(
    index='Product line', 
    columns='Branch', 
    values='gross income',
    aggfunc='count'
    ).round(1)
print(invoices_by_category)
```

```python
el_acc = invoices_by_category.loc['Electronic accessories']

max_el = el_acc.max()
max_br = el_acc.idxmax()

print(f'–ù–∞–π–±—ñ–ª—å—à–µ –ø–æ–∫—É–ø–æ–∫ –∑ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó Electronic accessories —î –≤ –ú–∞–≥–∞–∑–∏–Ω—ñ {max_br} = {max_el}')

```

```python
grouped_customers = (
    df.groupby(['Gender', 'Product line'])
    .size()
    .rename('Count')
    .groupby(level=0, group_keys=False)
    .apply(lambda x: x.sort_values(ascending=False))
)

print(grouped_customers)
```

```python
m_max = grouped_customers['Male'].idxmax()
m_min = grouped_customers['Male'].idxmin()

f_max = grouped_customers['Female'].idxmax()
f_min = grouped_customers['Female'].idxmin()

print(f'–ß–æ–ª–æ–≤—ñ–∫–∏ –Ω–∞–π–±—ñ–ª—å—à–µ –∫—É–ø—É—é—Ç—å {m_max}, –∞ –Ω–∞–π–º–µ–Ω—à–µ {m_min}')
print(f'–ñ—ñ–Ω–∫–∏ –Ω–∞–π–±—ñ–ª—å—à–µ –∫—É–ø—É—é—Ç—å {f_max}, –∞ –Ω–∞–π–º–µ–Ω—à–µ {f_min}')
```
