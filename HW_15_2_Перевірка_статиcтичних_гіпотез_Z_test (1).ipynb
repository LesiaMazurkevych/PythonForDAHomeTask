```python
#Умови

sample_mean = np.mean(data) 
print('sample_mean = ', sample_mean)

population_mean=20
print('population_mean = ', population_mean) 

population_std = np.std(data, ddof=1)
print('population_std = ', population_std)

sample_size = len(data)
print('sample_size = ', sample_size)

alpha = 0.05
print('alpha = ', alpha) 
```

```python
#Гіпотези

H₀: μ = 20 (середній чек)  
H₁: μ > 20 (підвищити середній чек)  
Тип тесту: правосторонній (one-tailed)  
```

```python
#SE (стандартне відхилення вибіркового середнього)

SE = population_std/np.sqrt(sample_size)
print('SE = ', SE)
```

```python
#Z-stat: z = (x̄ - μ₀) / SE

z_statistic = (sample_mean - population_mean) / SE
print('z_statisticSE = ', z_statistic)
```

```python
# Критичне значення
z_critical = stats.norm.ppf(1 - alpha)
print('z_critical = ', z_critical)
```

```python
# p_value 

p_value = 1 - stats.norm.cdf(z_statistic)
print('p_value = ', p_value)
```

```python
Порівняння z-статистики з критичним значенням:
z_statisticSE =  -0.995
z_critical =  1.645
z_statisticSE < z_critical -> Ho не відхиляємо

Аналіз p-value:
alpha = 0.05
p_value =  0.84 
p_value > alpha -> імовірність досить висока, тому Ho не відхиляємо

Висновок: статистично значущого підвищення середнього чеку за липень не спостерігається.
Фактично, середній чек навіть трохи знизився у порівнянні з 20$.
```
